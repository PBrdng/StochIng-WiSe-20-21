# Gesetz der grossen Zahlen

Wir untersuchen das Gesetz der grossen Zahlen in zwei Simulationen


---
### Gleichverteilung auf {1,...,15}

Sei $X \sim \mathrm{Unif}(\{1,\ldots,15\})$. 

Wir generieren $N = 10^5$ Zufallszahlen von $X$ und betrachten die Mittelwerte

```r
n = 15
N = 10^5
D = sample(1:n, N, replace=TRUE)
mean(D)
```

Als nächstes betrachten wir die Mittelwerte für verschiedene Werte von $N$ zwischen $1$ und $10^6$.

```r
Ns = seq(1, N, by = 100)

M = rep(0, length(Ns))
for (i in 1:length(Ns)){
    D = sample(1:n, Ns[i], replace=TRUE)
    M[i] = mean(D)
}
```

```r
plot(Ns, M, ylim = c(7,9), type = "l")
lines(Ns, rep(8, length(Ns)), col = "red", lw = 3)
```

---
### Geometrische Verteilung mit Parameter $p=0.4$

Sei $X \sim \mathrm{Geo}(0.4)$. 

Wir führen dasselbe Experiment wie oben durch.

```r
Ns = seq(1, N, by = 1000)
p = 0.4

M = rep(0, length(Ns))
for (i in 1:length(Ns)){
    D = rgeom(Ns[i], p) + 1 
    # +1, weil R die geometrische Verteilung als Anzahl Fehlversuche definiert. 
    # Wir definieren es als Anzahl bis der erste Erfolg kommt
    M[i] = mean(D)
}
```

```r
plot(Ns, M, ylim = c(2,3), type="l")
lines(Ns, rep(1/p, length(Ns)), col = "red", lw = 1.5)
```
